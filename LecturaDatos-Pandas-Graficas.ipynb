```python
# Distribución de Outcome
outcome = df['Outcome'].value_counts()
outcome.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Distribución de pacientes según diagnóstico de diabetes')
plt.xlabel('Outcome (0 = No, 1 = Sí)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
La gráfica muestra que la mayoría de los participantes no tienen diabetes, mientras que un menor porcentaje sí tiene diabetes. 
```

```python
# Histograma de Glucose
plt.hist(df['Glucose'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Niveles de Glucosa')
plt.xlabel('Glucose')
plt.ylabel('Frecuencia')
plt.show()
```

```python
La gráfica muestra que la mayoría de los pacientes tienen el nivel de la glucosa entre los valores del medio, mientras que pocos tienen niveles muy bajos o muy altos.
```

```python
# Boxplot del IMC
plt.boxplot(df['BMI'].dropna())
plt.title('Boxplot del Índice de Masa Corporal (BMI)')
plt.ylabel('BMI')
plt.show()
```

```python
El diagrama muestra que la mayoría de los valores del BMI estan ubicados entre 25 y 35, mientras que existen algunos valores fuera de lo comun ubicados por encima de 50, los cuales representan casos de obesidad extrema.
```

```python
# Promedio de Glucose según diagnóstico de diabetes
plt.figure(figsize=(6,5))
sns.boxplot(x='Outcome', y='Glucose', data=df, hue='Outcome', dodge=False, palette='pastel', legend=False)
sns.stripplot(x='Outcome', y='Glucose', data=df, color='orange', alpha=0.6)
plt.title('Niveles de Glucosa de acuerdo al diagnóstico de diabetes')
plt.xlabel('Outcome (0 = No, 1 = Sí)')
plt.ylabel('Glucose')
plt.show()
```

```python
La grafica mmuestra que las personas que fueron diagnosticados con diabetes tienen un nivel de glucosas mas alto, mientras que las personas que no fueron diagnosticados con diabetes tenian niveles mas bajos.
```

```python
# Promedio de BMI según diagnóstico de diabetes
prom_bmi = df.groupby('Outcome')['BMI'].mean()
prom_bmi.plot(kind='bar', color=['lightcoral', 'gold'])
plt.title('Promedio de BMI por Categoría de Outcome')
plt.xlabel('Outcome (0 = No, 1 = Sí)')
plt.ylabel('Promedio de BMI')
plt.show()
```

```python
La gráfica muestra que las personas con diabetes tienen un promedio de BMI más alto que quienes no la tienen.
```

```python
variables_numericas = df[['Glucose', 'BMI', 'Outcome']]
matriz_correlacion = variables_numericas.corr().round(2)
```

```python
matriz_correlacion
```

```python
La tabla muestra que el nivel de glucosa está más relacionado con tener diabetes, mientras que el BMI también influye, pero a una nivel mas baja.
```

```python
# Mapa de calor de correlación entre las variables principales
correlacion = df[['Glucose', 'BMI', 'Outcome']].corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación (Glucose, BMI, Outcome)')
plt.show()
```

```python
El mapa de calor demuestra que hay una conexión normal entre la glucosa y el resultado de diabetes, mientras que el IMC muestra una relación baja con esta condición.
```
